##Questiom 2 : What is a p-value, and how is it used in hypothesis testing? What does it mean if the p-value is very small (e.g., 0.01)

A p-value (probability value) is a statistical measure that indicates the likelihood of observing the data, or something more extreme, assuming the null hypothesis (
𝐻
0) is true. It quantifies the evidence against
𝐻
0.

Formally, the p-value is the probability of obtaining a test statistic at least as extreme as the one observed, given that the null hypothesis is true.


**How is the P-Value Used in Hypothesis Testing?**

1. **Compare the P-Value to the Significance Level (𝛼):**
        The significance level (𝛼) is a threshold, commonly set at 0.05 (5%), that determines the cut-off for rejecting 𝐻0.
        If p≤α: **Reject the null hypothesis** (H0). This suggests the observed result is statistically significant.
        If p>α: **Fail to reject the null hypothesis**. There is insufficient evidence to conclude that the effect or difference is real.
2. **Interpret Evidence:**

         Smaller p-values indicate stronger evidence against 𝐻0, suggesting that the observed result is unlikely to have occurred by random chance under 𝐻0.
         Larger p-values indicate weaker evidence against 𝐻0.

**What Does a Very Small P-Value Mean (e.g., 0.01)?**

  1. A p-value of 0.01 implies that there is only a 1% probability of observing the test statistic, or something more extreme, if 𝐻0 is true
  2. This suggests strong evidence against the null hypothesis. In most cases, researchers would reject 𝐻0 at a typicalα=0.05 orα=0.01 level.


**Key Points:**

1. Very Small P-Values:
         Indicate strong evidence against 𝐻0.
        Suggest the observed result is unlikely due to chance alone.

2. Limitations:
          A small p-value does not measure the size or importance of an effect, only its statistical significance.
          Results can still be practically insignificant despite being statistically significant.
3. Contextual Significance:
        Always consider the practical implications alongside the statistical results.

##question 7 Explain the Central Limit Theorem (CLT) with example.

The Central Limit Theorem (CLT) states that, for a sufficiently large sample size, the sampling distribution of the sample mean of a random variable will approximate a normal distribution (Gaussian distribution), regardless of the shape of the original population distribution. This approximation improves as the sample size increases.

**Key Points of CLT:**

1. Applies to the Sample Mean: It is the distribution of the sample mean (ˉX) that becomes normal.
2. Independence: The samples must be independent and identically distributed (i.i.d.).
3. Large Sample Size: As the sample size n increases, the approximation to normality improves.
4. Mean and Variance of Sampling Distribution:

          The mean of the sampling distribution is equal to the population mean:

**Example to Illustrate CLT**

Scenario:

Consider a population of exam scores that is uniformly distributed between 0 and 100 (not normal).

Steps:

1. **Population Distribution:** The population has a uniform distribution, meaning all scores between 0 and 100 are equally likely.
2. **Draw Random Samples:**
            a. Take random samples of size n=5 and compute their means.
            b. Repeat this process many times to create a sampling distribution of the mean.
3. **Observation:** If you plot the means of these samples, the shape of the sampling distribution begins to resemble a normal distribution, even though the original population is uniform.
4. **Increase Sample Size:** Repeat the process with a larger sample size, say n=30. The sampling distribution becomes more tightly clustered around the population mean and more closely approximates a normal distribution.


**Applications of CLT**

1. **Confidence Intervals:** Used to calculate confidence intervals for population parameters.
2. **Hypothesis Testing:** Assumes normality in the sampling distribution for test statistics.
3. **Real-World Scenarios:**
         a. **Manufacturing:** Analyzing the average weight of products in a batch.
         b. **Marketing:** Estimating average customer spending from sample data.
         c. **Healthcare:** Determining the average effect of a new drug based on clinical trial results.

## Question 9  : Given a sample mean of 105, a population mean of 100, a standard deviation of 15, and a sample size of 25, calculate the z-score and p-value. Based on a significance level of 0.05, do you reject or fail to reject the null hypothesis? Task: Write Python code to calculate the z-score and p-value for the given data.Objective: Apply the formula for the z-score and interpret the p-value for hypothesis testing.

```python
import scipy.stats as stats
import math

# Given data
sample_mean = 105
population_mean = 100
standard_deviation = 15
sample_size = 25
significance_level = 0.05

# Calculate z-score
z_score = (sample_mean - population_mean) / (standard_deviation / math.sqrt(sample_size))

# Calculate p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test

# Determine whether to reject the null hypothesis
reject_null = p_value < significance_level

# Output results
print(f"Z-Score: {z_score:.2f}")
print(f"P-Value: {p_value:.4f}")
if reject_null:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

##Question10: Simulate a binomial distribution with 10 trials and a probability of success of 0.6 using Python.Generate 1,000 samples and plot the distribution. What is the expected mean and variance? Task: Use Python to generate the data, plot the distribution, and calculate the mean and variance.Objective: Understand the properties of a binomial distribution and verify them through simulation.




```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the binomial distribution
n_trials = 10
p_success = 0.6
n_samples = 1000

# Simulate the binomial distribution
samples = np.random.binomial(n_trials, p_success, n_samples)

# Calculate the mean and variance of the samples
mean_simulated = np.mean(samples)
variance_simulated = np.var(samples)

# Theoretical mean and variance
mean_theoretical = n_trials * p_success
variance_theoretical = n_trials * p_success * (1 - p_success)

# Plot the distribution
plt.hist(samples, bins=range(n_trials + 2), density=True, alpha=0.75, color='blue', edgecolor='black')
plt.title('Binomial Distribution (n=10, p=0.6)')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.xticks(range(n_trials + 1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Output the results
print(f"Simulated Mean: {mean_simulated:.2f}")
print(f"Simulated Variance: {variance_simulated:.2f}")
print(f"Theoretical Mean: {mean_theoretical:.2f}")
print(f"Theoretical Variance: {variance_theoretical:.2f}")
```
